#2. Create DataBase for the Waste Food Management

```python
conn= sqlite3.connect("food_wastage.db", timeout= 10)
cursor= conn.cursor()
```

```python
# Creating table: providers
cursor.execute("""
CREATE TABLE IF NOT EXISTS providers (
    Provider_ID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Type TEXT,
    Address TEXT,
    City TEXT,
    Contact TEXT
)
""")
conn.commit()
```

```python
# Insert data using iterrow()
for index, row in providers.iterrows():
    cursor.execute(
        """ INSERT INTO providers (Provider_ID, Name, Type, Address, City, Contact)
            VALUES (?, ?, ?, ?, ?, ?)""",
        tuple(row))
    conn.commit()
```

###Create recievers dataset in DB

```python
# Creating table: receivers
cursor.execute("""
CREATE TABLE IF NOT EXISTS receivers (
    Receiver_ID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Type TEXT,
    City TEXT,
    Contact TEXT
)
""")
conn.commit()
```

```python
# Insert data using iterrow()
for index, row in receivers.iterrows():
    cursor.execute(
        """ INSERT INTO receivers (Receiver_ID, Name, Type, City, Contact)
            VALUES (?, ?, ?, ?, ?)""",
        tuple(row))
    conn.commit()
```

###To create food list in DB

```python
# Creating table: food_list
cursor.execute("""
CREATE TABLE IF NOT EXISTS food_list (
    Food_ID INTEGER PRIMARY KEY,
    Food_Name TEXT,
    Quantity INTEGER,
    Expire_Date DATE,
    Provider_ID INTEGER,
    Provider_Type TEXT,
    Location TEXT,
    Food_Type TEXT,
    Meal_Type TEXT,
    FOREIGN KEY (Provider_ID) REFERENCES providers(Provider_ID)
)
""")
conn.commit()
```

```python
# Insert data using iterrows()

for index, row in food_list.iterrows():
    cursor.execute(
        """ INSERT INTO food_list(Food_ID, Food_Name, Quantity, Expire_Date,
         Provider_ID, Provider_Type, Location, Food_Type, Meal_Type)
            VALUES (?, ?, ?, ?, ?,?,?,?,?)""",
        tuple(row))
```

### To create claims dataset in DB

```python
# Creating table: claims
cursor.execute("""
CREATE TABLE IF NOT EXISTS claims (
    claim_id SERIAL PRIMARY KEY,
    Food_ID INTEGER,
    Receiver_ID INTEGER,
    Status TEXT,
    Timestamp TIMESTAMP,
    FOREIGN KEY (Food_ID) REFERENCES food_lists(Food_ID),
    FOREIGN KEY (Receiver_ID) REFERENCES receivers(Receiver_ID)
)
""")
conn.commit()
```

```python
# Insert data using iterrows()

for index, row in claims.iterrows():
    cursor.execute(
        """ INSERT INTO claims(Claim_ID, Food_ID, Receiver_ID, Status, Timestamp)
            VALUES (?, ?, ?, ?,?)""",
        tuple(row))
```

#3. General Queries

###1. Which city has the highest number of food providers?

```python
# SQL query to get the city that provides the hightest numbers of food
query= """
SELECT City, COUNT(*) AS count_providers
FROM providers
GROUP BY city
ORDER BY count_providers DESC
LIMIT(1);
"""

cursor.execute(query)
result= cursor.fetchall()

# Convert result into a Dataframe for the better readibility
df= pd.DataFrame(result, columns= ['City', 'count_providers'])
df
```

###2. Which type of food provider (restaurant, grocery store, etc.) contributes the most food?


```python
query = """
SELECT provider_type, COUNT(provider_type) AS count_provider_type
FROM food_list
GROUP BY provider_type
ORDER BY  count_provider_type DESC
LIMIT(3);
"""
cursor.execute(query)
result= cursor.fetchall()

# Convert result into a Dataframe for the better readibility
df= pd.DataFrame(result, columns= ['provider_type', 'count_provider_type'])
df
```

###3. What is the contact information of food providers in a specific city?


```python
query = """
SELECT DISTINCT city FROM providers;
"""
cursor.execute(query)
result= cursor.fetchall()

# Convert result into a Dataframe for the better readibility
df= pd.DataFrame(result, columns= ['city'])
df
```

###4. Which receivers have claimed the most food?


```python
query= """
SELECT
    r.Name AS Receiver_Name,
    COUNT(c.Claim_ID) AS Total_Claims
FROM claims AS c
JOIN receivers AS r
ON c.Receiver_ID = r.Receiver_ID
WHERE c.Status = 'Completed'
GROUP BY r.Name
ORDER BY Total_Claims DESC
LIMIT(3);
"""
cursor.execute(query)
result= cursor.fetchall()

# Convert result into a Dataframe for the better readibility
df= pd.DataFrame(result, columns= ['Receiver_Name', 'Total_Claims'])
df
```

##Food Listings & Availability

###5. What is the total quantity of food available from all providers?

```python
query = """
SELECT SUM(Quantity) AS Total_Quantity
FROM food_list;
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Total_Quantity'])
print(df)
```

###6. Which city has the highest number of food listings?

```python
query = """
SELECT Location, count(Location) AS Total_Food_listings
FROM food_list
GROUP BY Location
ORDER BY Total_Food_listings DESC
LIMIT(1);
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Location', 'Total_Food_Listings'])
print(df)
```

###7. What are the most commonly available food types?

```python
query = """
SELECT Food_Type, COUNT(Food_Type) AS count_food_type
FROM food_list
GROUP BY Food_Type
ORDER BY count_food_type DESC;
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Food_Type', 'count_food_type'])
print(df)
```

##Claims & Distribution


 ### 8. How many food claims have been made for each food item?

```python
query = """
SELECT fl.Food_Type, COUNT(Claim_ID) AS total_claim
FROM claims AS c
JOIN food_list AS  fl
ON c.Food_ID = fl.Food_ID
GROUP BY fl.Food_type
ORDER BY total_claim DESC;
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Food_Type', 'total_claim'])
print(df)
```

### 9. Which provider has had the highest number of successful food claims?

```python
query = """
SELECT
	p.Name,
	COUNT(c.Status) AS total_claim
FROM food_list AS fl
JOIN  providers AS p
ON  fl.Provider_ID =p.Provider_ID
JOIN claims AS c
ON fl.Food_ID = c.Food_ID
WHERE c.Status = 'Completed'
GROUP BY p.name
ORDER BY total_claim DESC
LIMIT(1);
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Name', 'total_claim'])
print(df)
```

###   10. What percentage of food claims are completed vs. pending vs. canceled?


```python
query = """
SELECT
    Status,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM claims), 2) AS percentage
FROM claims
GROUP BY Status
ORDER BY percentage;
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Status', 'parcentage'])
print(df)
```

##Analysis & Insights

###11. What is the average quantity of food claimed per receiver?

```python
query = """
SELECT r.Name, ROUND(AVG(fl.Quantity), 2) AS avg_quantity
FROM claims AS c
JOIN receivers AS r
ON c.Receiver_ID = r.Receiver_ID
JOIN food_list AS fl
ON c.Food_ID = fl.Food_ID
GROUP BY r.Name
ORDER BY avg_quantity DESC;
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Name', 'avg_quantity'])
print(df)
```

###12. Which meal type (breakfast, lunch, dinner, snacks) is claimed the most?


```python
query = """
SELECT fl.Meal_Type, COUNT(c.Claim_ID) AS count_meal_type
FROM claims AS c
JOIN food_list AS fl
ON c.Food_ID = fl.Food_ID
GROUP BY fl.Meal_Type
ORDER BY count_meal_type DESC;
"""
cursor.execute(query)
result = cursor.fetchall()
# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Meal_Type', 'count_meal_type'])
print(df)

```


###13. What is the total quantity of food donated by each provider?

```python
query = """
SELECT p.name, SUM(fl.Quantity) AS total_donation
FROM food_list as fl
JOIN providers as p
ON fl.Provider_ID = p.Provider_ID
GROUP BY p.Name
ORDER BY total_donation DESC;
"""
cursor.execute(query)
result = cursor.fetchall()

# Convert result into a DataFrame for better readability
df = pd.DataFrame(result, columns=['Name', 'total_donation'])
print(df)
```

###4. Products and quantities by expire dates

```python
query= """
SELECT
    Food_Type,
	Meal_Type,
    Quantity,
    Expire_Date,
    (Expire_Date - CURRENT_DATE) AS Days_To_Expire
FROM food_list
ORDER BY Expire_Date ASC
LIMIT 40;
"""
cursor.execute(query)
result= cursor.fetchall()
# Convert result into a DataFrame for better readibility
df= pd.DataFrame(result, columns= ['Food_Type', 'Meal_Type', 'Quantity', 'Expire_Date', 'Days_To_Expire'])
df
```
